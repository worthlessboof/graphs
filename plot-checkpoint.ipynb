```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.legend import Legend
%matplotlib inline
plt.rc('axes',edgecolor='black')
```

```python
labels = ['Obs', 'Obs', 'Model']
red = np.array([0.8, 0, 1])
red_hatch = np.array([0, 1.3, 0])
blue_hatch = np.array([0, 4.5, 0.5])
light_blue = np.array([0, 0, 2])
blue = np.array([5, 0, 0])

ind = [x for x, _ in enumerate(labels)]

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_ylim(0, 7)
r=ax.bar(ind, red, width=0.8, label='POA', color='red')
rh=ax.bar(ind, red_hatch, width=0.8, label='Insoluble',color='none', edgecolor='red', hatch="/////", bottom=red)
bh=ax.bar(ind, blue_hatch, width=0.8, label='WSQA',color='none', edgecolor='Blue', hatch="/////", bottom=red+red_hatch)
lb=ax.bar(ind, light_blue, width=0.8, label='Other SOA', color='#AAAAFF', bottom=red+red_hatch+blue_hatch)
b=ax.bar(ind, blue, width=0.8, label='SOA', color='Blue', bottom=red+red_hatch+blue_hatch+light_blue)

plt.xticks(ind, labels)
plt.ylabel("SOAS-CTR")
leg1 = Legend(ax, [b, r], ["SOA", "POA"], loc='upper left', edgecolor = "black")
leg2 = Legend(ax, [bh, rh], ["WSQA", "Insoluble"], loc='upper center', edgecolor = "black")
leg3 = Legend(ax, [lb, bh, r], ["Other SOA", "Aq. SOA", "POA"], loc='upper right', edgecolor = "black")
ax.add_artist(leg1);
ax.add_artist(leg2);
ax.add_artist(leg3);

#plt.show()
```
